```python
from functools import reduce 
import pandas as pd 
import os 
```

```python
carpeta = r"C:\my_last_dance\final_provider\final_provider" 
dataframes_dict = {}  # Diccionario para almacenar los DataFrames

for archivo in os.listdir(carpeta):
    ruta_completa = os.path.join(carpeta, archivo)

    if archivo.endswith(".csv"):
        nombre_variable = archivo.replace(".csv", "")  # Eliminar la extensión para usar como clave
        dataframes_dict[nombre_variable] = pd.read_csv(ruta_completa) 
# Acceder a un DataFrame específico

dataframes_dict['phone'].info()
print()
dataframes_dict['personal'].info()
print()
dataframes_dict['contract'].info()
print()
dataframes_dict['internet'].info()
```

```python
grouped_df = reduce(lambda left, right: pd.merge(left, right, on="customerID", how="outer"), dataframes_dict.values())
print(grouped_df.info())

```
